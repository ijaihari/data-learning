## Filtering

```python
data = {
    "Product": ["Laptop", "Laptop", "Mobile", "Mobile", "Tablet", "Tablet", "Headphones", "Camera"],
    "Category": ["Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Electronics", "Accessories", "Electronics"],
    "Brand": ["Dell", "HP", "Samsung", "Apple", "Lenovo", "Apple", "Sony", "Canon"],
    "Price": [55000, 62000, 18000, 70000, 25000, 30000, 4000, 45000],
    "Units_Sold": [120, 80, 150, 110, 90, 60, 200, 70],
    "City": ["Mumbai", "Delhi", "Mumbai", "Chennai", "Bangalore", "Hyderabad", "Mumbai", "Delhi"],
    "Date": [
        "2024-01-05", "2024-01-10", "2024-02-15", "2024-02-20",
        "2024-03-01", "2024-03-05", "2024-03-15", "2024-03-20"
    ]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
df.sort_values('Date')
```

```python
m = df['Date'].dt.month
m
```

```python
type(m)
```

```python
df['Week'] = df['Date'].dt.isocalendar().week
```

```python
df
```

```python
df['Month_Name'] = df['Date'].dt.month_name()
df
```

```python
df['Day_Name'] = df['Date'].dt.day_name()
df
```

```python
df['Month_Start'] = df['Date'].dt.to_period('M').dt.start_time
df
```

```python
m = df['Date'].dt.to_period('D')
m
```

```python
new_db = pd.DataFrame(m)
```

```python
new_db.to_csv('Date.csv')
```
