```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url_2024 = "https://www.elections.tn.gov.in/ACwise_Gendercount_01012024.aspx"

df_2024 = pd.read_html(url_2024, header=0)[0]
df_2024
```

```python
df_2024_ssr = pd.Dataframe()
df_2024_ssr.columns = [
    "AC_Info",
    "Male_2024",
    "Female_2024",
    "ThirdGender_2024",
    "Total_2024"
]
df_2024_ssr[["AC_No", "AC_Name"]] = df_2024["AC_Info"].str.split(" ", n=1, expand=True)
df_2024_ssr["AC_No"] = df_2024_ssr["AC_No"].astype(int)
df_2024_ssr.drop(columns=["AC_Info"], inplace=True)
df_2024_ssr.head()
```
