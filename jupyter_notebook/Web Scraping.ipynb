```python
#The below url contains an html table with data about colors and color codes.
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/HTMLColorCodes.html"
```

```python
data  = requests.get(url).text
data
```

```python
soup = BeautifulSoup(data,"html.parser")
table = soup.find('table')
soup
```

```python
for row in table.find_all('tr'): # in html table row is represented by the tag <tr>
    # Get all columns in each row.
    cols = row.find_all('td') 
    color_name = cols[2].string # store the value in column 3 as color_name
    color_code = cols[3].string # store the value in column 4 as color_code
    print("{}--->{}".format(color_name,color_code))
```

```python
with open('person.json', 'w') as f:
    json.dump(person, f)
```

```python
json_object = json.dumps(person, indent = 4) 
json_object
```

# Create and Writing XML

```python
import xml.etree.ElementTree as ET

root = ET.Element("students")  # root tag

student1 = ET.SubElement(root, "student")
ET.SubElement(student1, "name").text = "Jai"
ET.SubElement(student1, "age").text = "21"
ET.SubElement(student1, "course").text = "Data Science"

student2 = ET.SubElement(root, "student")
ET.SubElement(student2, "name").text = "Hari"
ET.SubElement(student2, "age").text = "22"
ET.SubElement(student2, "course").text = "AI & ML"

tree = ET.ElementTree(root)
tree.write("students.xml", encoding="utf-8", xml_declaration=True)
```

# Reading XML Files

```python
import xml.etree.ElementTree as ET

tree = ET.parse("students.xml")
root = tree.getroot()

for student in root.findall("student"):
    name = student.find("name").text
    age = student.find("age").text
    course = student.find("course").text
    print(name, age, course)
```
