```python
import pandas as pd
import seaborn as sns
import csv, sqlite3
```

```python
conn = sqlite3.connect("socioeconomic.db")
```

```python
df = pd.read_csv('https://data.cityofchicago.org/resource/jcxq-k9xf.csv')
df.to_sql("chicago_socioeconomic_data", conn, if_exists ='replace', index = False)
```

```python
query = "select count(*) from chicago_socioeconomic_data"
df = pd.read_sql(query, conn)
df
```

```python
query = "select count(*) from chicago_socioeconomic_data where hardship_index > 50.0"
pd.read_sql(query, conn)
```

```python
query = "select MAX(hardship_index) from chicago_socioeconomic_data"
pd.read_sql(query, conn)
```

```python
query  = "select community_area_name from chicago_socioeconomic_data where hardship_index = ( select max(hardship_index) from chicago_socioeconomic_data );"
pd.read_sql(query, conn)
```

```python
query  = "SELECT community_area_name FROM chicago_socioeconomic_data WHERE per_capita_income_ > 60000;"
pd.read_sql(query, conn)
```
