```python
import pandas as pd
from scipy.stats import chi2_contingency

data = [[40, 60],
        [10, 90]]

table = pd.DataFrame(
    data,
    index=["Smoker", "Non-Smoker"],
    columns=["Cancer Yes", "Cancer No"]
)

chi2, p, dof, expected = chi2_contingency(table)
```
