```python
import pandas as pd

df = pd.DataFrame({
    'City': ['Mumbai', 'Delhi', 'Mumbai', 'Chennai', 'Delhi', 'Mumbai'],
    'Sales': [100, 150, 200, 120, 180, 160]
})

df
```

```python
df['City'].dtype
```

```python
df['City'].memory_usage(deep=True)
```

```python
df['City'] = df['City'].astype('category')
```

```python
df['City'].memory_usage(deep=True)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Age': [22, 25, 47, 52, 46],
    'Salary': [25000, 32000, 58000, 72000, 60000]
})

df
```
